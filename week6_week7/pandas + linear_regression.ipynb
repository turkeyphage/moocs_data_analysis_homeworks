```python
# 取出最後三百期的資料   (10分)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#最後三百期的資料
last_300 = df.tail(300)
last_300
```

```python
# 從這三百期資料，取出收盤價close，並轉成numpy_array讓我們在後面使用   （10分）
```

```python

#取出收盤價close
df_close = last_300['Close']
#轉成numpy_array
df_close = df_close.values
df_close
```

```python
# 我們現在要做的，是用前一天的收盤價，預設第二天的收盤價（星期一預測星期二），
# 請依照這個需求，把資料分成題目X(迴歸直線的x座標，表示所有的「前一天收盤價」)
# 跟答案Y(迴歸直線的y座標，表示所有的「第二天收盤價」)   （10分）
```

```python
#題目X
X = df_close[:-1]
#答案Y
Y = df_close[1:]
```

```python
# 請把題目X，答案Y再分成佔80%的訓練資料跟佔20%的測試資料（10分）
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state = 87)
```

```python
# 請實際的利用訓練資料和scikit learn跑線性回歸，畫圖，並在得到的迴歸直線上，
# 加上訓練資料的資料點(plt.scatter...)，讓大家看看你回歸的結果跟實際結果差多少。   (10分)
```

```python


x_train = x_train.reshape(len(x_train), 1)
x_test = x_test.reshape(len(x_test), 1)

from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train, y_train)

#train data
plt.subplot(1 , 2, 1)
plt.scatter(x_train, y_train)
plt.plot(x_train, regr.predict(x_train), 'r')


#test data
plt.subplot(1, 2, 2)
plt.scatter(x_test, y_test)
plt.plot(x_test, regr.predict(x_test), 'r')

```
