# 第一題：
題目：給定一個矩陣m，裡面的值都大於0。
請將矩陣m裡大於等於1的數字都變成1，小於1的變成0。
 
m = np.array([[0,1,10],[2,6,7]])
 
提示：使用巢狀迴圈(nested loop)加判斷式

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

m = np.array([[0.4, 1.5, 10], [1.0, 0.254, 2.8]])
print("original matrix =")
print(m)

filtedList = list()
for x in m:
    row_list = list()
    for y in x:
        if y >= 1:
            row_list.append(1)
        else:
            row_list.append(0)
    filtedList.append(row_list)
m = np.array(filtedList)

print("after transforming:")
print(m)
```

# 第二題：
有A,B,C三個班級，每班皆有10個學生，期中考結束後，老師希望計算各班級每位學生的加權平均成績。
 
請隨機產生3個班級的成績單,每班10人,共五科,權重為weights=[0.25,0.25,0.25,0.15,0.10]。
請算出班級加權平均成績，並以一個array輸出。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

A = np.random.randint(101, size=(10,5))
B = np.random.randint(101, size=(10,5))
C = np.random.randint(101, size=(10,5))
print("A班成績 = \n{}\n".format(A))
print("B班成績 = \n{}\n".format(B))
print("C班成績 = \n{}\n".format(C))

weights=np.array([0.25,0.25,0.25,0.15,0.10])
a_dot = np.dot(A, weights)
b_dot = np.dot(B, weights)
c_dot = np.dot(C, weights)

print("A班成績加權後 = {}\n".format(a_dot))
print("B班成績加權後 = {}\n".format(b_dot))
print("C班成績加權後 = {}\n".format(c_dot))


print("A,B,C 三個班級加權平均成績各自為 {}".format(np.array([np.average(a_dot), np.average(b_dot), np.average(c_dot)])))
```

# 挑戰題一：
請定義一個全班高微成績的函式，輸入是一個(3, 30, 3)的array，  
表示3個班級中30位學生的成績(小考，期中考，期末考)  。

目標輸出則是一個(3, 30, 1)的array，表示所有學生的期末成績，
期末成績計算方式如下：

1. 小考20%，期中考35%，期末考45%  
2. 期中考40%，期末考60%。  
擇優作為期末成績
 
Hint
* 可以用雙重for迴圈 分別跑過班級跟人
* 可以先設一個array裝算出來的成績
* 記得要用return 
* 請做個測試，檢驗你的結果，方便同學評分。


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

students_grades = np.random.randint(101, size=(1, 3*30*3)).reshape(3,30,3)

print("原始成績：\n{}".format(students_grades))

def count_final(all_grades):
    weights1 = np.array([0.2, 0.35, 0.45])
    weights2 = np.array([0.4, 0.6])
    
    final_grades = list()
    
    for classNum in all_grades:
        for personGrades in classNum:
            #除去小考的成績
            no_pratices_grades = personGrades[1:]

            #以weights1計算
            after_weights1 = np.dot(personGrades, weights1)
            
            #以weights2計算
            after_weights2 = np.dot(no_pratices_grades, weights2)
            
            if after_weights1 >= after_weights2:
                final_grades.append(after_weights1)
            else:
                final_grades.append(after_weights2)
    
    return np.array(final_grades).reshape(3, 30, 1)


print("\n期末成績：\n{}".format(count_final(students_grades)))
```
