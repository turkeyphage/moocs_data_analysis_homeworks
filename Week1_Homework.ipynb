# 第二題：台北市最高最低溫分佈圖
題目說明：請以plot函式畫出台北市近年來的高低溫分佈圖

```python
month_name = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
month = range(1, 13)
highest_temp = [19.1, 19.6, 22.1, 25.7, 29.2, 32.0, 34.3, 33.8, 31.1, 27.5, 24.2, 20.7]
lowest_temp = [13.9, 14.2, 15.8, 19.0, 22.3, 24.6, 26.3, 26.1, 24.8, 22.3, 19.3, 15.6]
plot(month, highest_temp, 'r-', label = 'highest_temp')
plot(month, lowest_temp, 'b-', label = 'lowest_temp')
pyplot.xlabel("MONTH")
pyplot.ylabel(u'\N{DEGREE SIGN}'+ 'C')
pyplot.xticks(month, month_name)
pyplot.title("Temperature of Taipei City")
pyplot.legend(loc=1)
```

# 挑戰題一：買相機
團法人大兔兔文教基金會要添購攝影設備。
他們預計在未來四個月內分別添購四項價格為 100k,150k,230k,80k的設備，
請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例。

```python
#累積消費額
pyplot.figure(1)
month = [x+1 for x in range(0, 4)]
month_name = ['month'+ str(x) for x in month]
item_price = [100000, 150000, 230000, 80000]

def cumulative_sum(values, start=0):
    for v in values:
        start += v
        yield start
cumulative_cost_list = list(cumulative_sum(item_price))
plot(month, cumulative_cost_list)
pyplot.title("Cumulative Cost")
pyplot.xticks(month, month_name)
pyplot.xlabel("Month")
pyplot.ylabel("Total Cost")

#圓餅圖
pyplot.figure(2)
pyplot.title("Proportion of Costs")
explode = (0, 0.1, 0, 0)
pyplot.pie(item_price, explode=explode, labels=month_name, autopct='%1.1f%%', shadow=True, startangle=90)
pyplot.axis('equal')
```

# 挑戰題二：自己出！

題目：使用input 建立一個由日期查詢星期幾的系統

```python
from datetime import date
import calendar
usr_input = input("year/month/day (ex: 1982/08/13)= ")
year, month, day = usr_input.split("/")
searchDate = datetime.datetime.strptime(day+month+year, "%d%m%Y").date()
answer = calendar.day_name[searchDate.weekday()]


print('-'.join([year,month,day]) + ' is ' + answer)
```
